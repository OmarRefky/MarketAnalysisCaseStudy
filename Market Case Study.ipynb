#### 2.1.2. Outliers <a name='out'></a>

```python
# looking at the distribution of `order_price_before_discount` column
fig, ax = plt.subplots(1,2,figsize=(15,4))
sns.histplot(data=orders_df, x='order_price_before_discount', ax=ax[0])
sns.boxplot(data=orders_df, x='order_price_before_discount', ax=ax[1])
fig.suptitle('Orders price before discount Distributions')
plt.show()

fig, ax = plt.subplots(1,2,figsize=(15,4))
sns.histplot(data=orders_df, x='discount', ax=ax[0])
sns.boxplot(data=orders_df, x='discount', ax=ax[1])
fig.suptitle('Discount Distributions')
plt.show()
```

```python
outliers_indices = []
numerical_features = ['order_price_before_discount','discount']
for feature in numerical_features: # loop over numerical features
    Q3 = np.quantile(orders_df[orders_df.sales_order_status=='Delivered'][feature], 0.75) #calculate 3rd quartile 
    Q1 = np.quantile(orders_df[orders_df.sales_order_status=='Delivered'][feature], 0.25) #calculate 1st quartile
    IQR = Q3 - Q1 # Calculate Interquartile Range
    # get the indices of identified outliers and append them to the list
    # Q3 + 1.5(IQR) < outlier or outlier < Q1 - 1.5(IQR)
    outliers_indices.extend(orders_df[orders_df.sales_order_status=='Delivered'] \
                            [(orders_df[feature] > (Q3 + (1.5 * IQR))) | (orders_df[feature] < (Q1 - (1.5 * IQR)))].index)
outlier_indices_dict = Counter(outliers_indices) # creates a dict with the frequency of each index
# unpack the dict and flag the index as potential outlier if the freq is higher than 1
potential_outliers_indices = list(element for element, freq in outlier_indices_dict.items() if freq > 1)
print(f'Found {len(potential_outliers_indices)} potential outliers!')
```

```python
# display outliers and sort by price before discount then discount
outliers_df = orders_df.loc[potential_outliers_indices].sort_values(numerical_features, ascending=False)
display(outliers_df)
```

```python
# How much is 721 outliers out of the entire data
outliers_perc = round(len(potential_outliers_indices)/len(orders_df)*100,2)
print(f'Outliers raito to entire orders dataframe = {outliers_perc}%')
```

After looking into the percentage of the outliers and their entire records it doesnt make much sense to remove them, Specially since I wont be using linear modeling that can be skewed by outliers but its good to note they exist.

**Modeling**

```python
### choosing model hyper parameters
# loop over each training set
for dataset in [orders, revenues]:
    # create dataframe to store results
    results_df = pd.DataFrame(columns=['p','q','AIC','BIC'])

    # creating index for looping
    i = 0

    # Loop over p values from 0-5
    for p in range(6):
        # Loop over q values from 0-5
        for q in range(6):
        # avoid value errors
            try:
                # create and fit ARIMA(p,1,q) model
                model = ARIMA(dataset, order=(p,1,q))
                results = model.fit()

                # add the results to the dataframe
                results_df.loc[i,'p'] = p
                results_df.loc[i,'q'] = q
                results_df.loc[i,'AIC'] = results.aic
                results_df.loc[i,'BIC'] = results.bic

            except:
                results_df.loc[i,'p'] = p
                results_df.loc[i,'q'] = q
                results_df.loc[i,'AIC'] = None
                results_df.loc[i,'BIC'] = None

            # increament the index
            i += 1

    print(f'\n\nBest Predictive models for {dataset.columns[0]}',end='')
    display(results_df.sort_values('AIC').head())
```

The best predictive model for orders series is ARIMA(3,1,3) and for revenues series is ARIMA(3,1,5)

### 3.7. Considering the Stock snapshot and the sales data per SKU, Calculate days on hand for all SKUs (time remaining till SKUs run out of stock).<a name='stock1'></a> 


### Summary
- Cleaned duplicated rows from the data.
- Aggregated total sales per SKU
- Calculated average daily sales per SKU
- Calculated days on hand for per SKU


```python
# drop duplicated rows
inventory_df = sales_df.drop_duplicates()
inventory_df
```

```python
# Aggregating total sales per SKU and join it with stocks data
sku_sales = inventory_df.groupby('sku')['Sales_Count'].agg('sum').to_frame(name='Total_Sales_Count')\
                                                      .join(stocks_df.drop_duplicates().set_index('sku'), on='sku')

# calculate duration in days
duration = (inventory_df.Date.max() - inventory_df.Date.min()).days

# Calculate average daily sku sales
sku_sales['Avg_Daily_Sales'] = sku_sales.Total_Sales_Count.div(duration)

# Calculate days on hand for each SKU and floor it
sku_sales['Days_on_Hand'] = sku_sales.Stock.div(sku_sales.Avg_Daily_Sales).apply(np.floor).astype('int64')

# Present the output sorted from least to highest
stock_days_on_hand = sku_sales[['Stock','Days_on_Hand']].sort_values(['Days_on_Hand','Stock'])
display(stock_days_on_hand)

# save to csv
stock_days_on_hand.to_excel("stock_days_on_hand.xlsx", 'Stock_Days_on_Hand')
```

### 3.8. Calculate the stocks to be purchased from suppliers per SKU to reach 7 days on hand for all SKUs.<a name='stock2'></a> 


### Summary
- Filtered SKUs that will run out in less than 7 days.
- Calculated amount of stocks needed to be purchased per SKU.


```python
# filter out SKUs with more than 7 days on hand
sku_to_purchase = sku_sales[sku_sales.Days_on_Hand < 7]

# Calculate how many days left from days on hand to complete 7 days
sku_to_purchase['Days_to_Stock'] = 7 - sku_to_purchase.Days_on_Hand

# calculate amount to purchase and ceil the value
sku_to_purchase['Stock_to_Purchase'] = (sku_to_purchase.Days_to_Stock * sku_to_purchase.Avg_Daily_Sales).apply(np.ceil)

# present output
stocks_to_purchase = sku_to_purchase[['Stock','Days_on_Hand','Stock_to_Purchase']].sort_values(['Days_on_Hand','Stock'])
display(stocks_to_purchase)


# save to excel
stocks_to_purchase.to_excel("stocks_to_purchase.xlsx", 'Stocks_to_Purchase')
```
